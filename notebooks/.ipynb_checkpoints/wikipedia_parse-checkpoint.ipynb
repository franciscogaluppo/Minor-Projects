# Creating a graph of brazilian cities

Bla Bla

```python
import re
import requests
```

Bla Bla

```python
wiki = "https://pt.wikipedia.org"
list_link = "https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_do_Brasil_por_popula%C3%A7%C3%A3o"

regex_neighbors = '(?<=Municípios\ limítrofes</td><td\ style="vertical-align:\ top;\ text-align:\ left;">)(.*?)(?=</tr>)'
regex_href = "href=([\s\S]*?)title"
```

```python
html = requests.get(list_link).text.rstrip().replace("\n", "")
city_list = re.findall('<tr>([\s\S]*?)</tr>', html)[1:]
```

Bla Bla

```python
def parse_city(string):
    cp = re.findall('<td.*?>([\s\S]*?)</td>', string)
    aux = re.findall(regex_href, cp[2])[0].replace('"', '').split()[0]
    return [cp[1], wiki+aux]

parse_city(city_list[18])
```

Bla Bla

```python
lid = dict()
neighbors = dict()
error = list()
```

```python
def add_city(id, link):    
    parse = requests.get(link).text.rstrip().replace("\n", "")
    neigh = re.findall(regex_neighbors, parse)
    
    if len(neigh) > 0:
        lid[link] = id
        neighbors[id] = [wiki+x.replace('"', '').split()[0] for x in re.findall(regex_href, neigh[0])]
    elif [id, link] not in error:
        error.append([id, link])
```

Bla Bla

```python
j = 0
for i in city_list:
    if j % 100 == 0:
        print(j)
    j += 1
    
    id, link = parse_city(i)
    
    parse = requests.get(link).text.rstrip().replace("\n", "")
    neigh = re.findall(regex_neighbors, parse)
    
    if len(neigh) > 0:
        lid[link] = id
        neighbors[id] = [wiki+x.replace('"', '').split()[0] for x in re.findall(regex_href, neigh[0])]
    else:
        error.append([id, link])
```
