# Ergodicity

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Brownian Motion

```python
def bm(mu=0, sig=1, t=10, dt=0.01, init=0):
    sqrt_dt = np.sqrt(dt)
    x = [init]
    i = 0
    while i < t:
        i += dt
        x.append(x[-1] + mu*dt + sig*sqrt_dt*np.random.normal())
    return np.array(x)
```

```python
for i in range(100):
    plt.plot(bm(t=100))
plt.show()
```

### Geometric Brownian Motion

```python
def gbm(mu=0.5, sig=1, t=10, dt=0.01, init=1):
    sqrt_dt = np.sqrt(dt)
    x = [init]
    i = 0
    while i < t:
        i += dt
        x.append(x[-1] * (1 + mu*dt + sig*sqrt_dt*np.random.normal()))
    return np.array(x)
```

```python
for i in range(10):
    plt.plot(gbm(mu=0.1, sig=0.5, t=2))
plt.show()
```

### Time rate VS Ensamble rate

```python
def rates(fun, T=1000, N=1000, v=lambda x: x, init=10, dt=0.01):
    x = fun(t=T, dt=dt, init=init)
    try:
        time = v(np.power(x[-1]/x[0], 1/(len(x)-1))) / dt
    except:
        raise Exception("Imaginary rate.")
    
    x = [fun(t=dt, dt=dt, init=init)[-1] / init for i in range(N)]
    if any(y < 0.0 for y in x):
        raise Exception("Imaginary rate.")
        
    mean = v(np.mean(x)) / dt

    return (time, mean)
```

```python
def growth(fun, T=1000, v=lambda x: x, **kwargs):
    x = fun(t=T, **kwargs)
    return (v(x[-1]) - v(x[0])) / T
```

```python
growth(gbm, v=np.log, T=100, mu=2.6, sig=2, dt=0.01)
```
